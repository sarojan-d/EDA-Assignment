```python
# Heatmap (Categorical - Categorical)
heatmap_data = pd.crosstab((df['rating'])-(df['rating'])%1,(df['likedPercent'])/20-((df['likedPercent'])/20)%1)

sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Rating VS Liked Percentage')
plt.show()
```

```python
df['publishDate'] = pd.to_datetime(df['publishDate'])

# Line plot
sns.lineplot(x='publishDate', y='likedPercent', data=df)
plt.xlabel('Date')
plt.ylabel('Liked Percentage')
plt.show()
```

```python
# Convert 'Date' to datetime format
df['publishDate'] = pd.to_datetime(df['publishDate'],format='mixed')

# Extracting month and year from the 'publishDate' column
df['Month'] = df['publishDate'].dt.month
df['Year'] = df['publishDate'].dt.year

sns.heatmap(df.pivot_table(values='Rating',index='Month',columns='Year'))
```

```python
# Creating a pivot table
pivot_table = df.pivot_table(values='pages', index='Month', columns='Year', aggfunc='mean')

# Filling missing values with 0
pivot_table_filled = pivot_table.fillna(0)

# Using sns.clustermap directly on the filled pivot table
sns.clustermap(pivot_table_filled, cmap='YlGnBu', annot=True, fmt=".2f", linewidths=.5)
plt.title('Clustermap of Total Page Number Over Months and Years')
plt.show()
```
